```python
import pandas as pd
import re
import nltk
import csv
```

```python
tweets_streamed = pd.read_csv('AFGvsAUS.csv')
```

```python
tweets = pd.DataFrame(tweets_streamed['tweet_text'])
```

```python
def user_mention(tweet):
    return re.sub(r'@[\S]+','',tweet)

def convert_to_lowercase(tweet):
    return tweet.lower()

def url_handling(tweet):
    return re.sub(r'((www\.[\S]+)|(https?://[\S]+))','',tweet)

def RT_handling(tweet):
    return re.sub(r'\brt\b', '', tweet)

def dot_handling(tweet):
    return re.sub(r'\.{2,}', ' ', tweet)

def space_handling(tweet):
    return tweet.strip(' "\'')

def hashtag_handling(tweet):
    return re.sub(r'#(\S+)', r' \1 ', tweet)

def smile(tweet):
    return re.sub(r'(:\s?\)|:-\)|\(\s?:|\(-:|:\'\))',' EMO_POS ', tweet)

def laugh(tweet):
    return re.sub(r'(:\s?D|:-D|x-?D|X-?D)',' EMO_POS ', tweet)

def love(tweet):
    return re.sub(r'(<3|:\*)',' EMO_POS ', tweet)

def wink(tweet):
    return re.sub(r'(;-?\)|;-?D|\(-?;)',' EMO_POS ', tweet)

def sad(tweet):
    return re.sub(r'(:\s?\(|:-\(|\)\s?:|\)-:)',' EMO_NEG ', tweet)

def cry(tweet):
    return re.sub(r'(:,\(|:\'\(|:"\()',' EMO_NEG ', tweet)

#def clap(tweet):
    return re.sub(r'(*゜▽゜ノノ゛☆','EMO_POS', tweet)

def deEmojify(tweet):
    return tweet.encode('ascii', 'ignore').decode('ascii')
```

```python
tweets['tweet_text'] = tweets['tweet_text'].apply(user_mention)
tweets['tweet_text'] = tweets['tweet_text'].apply(convert_to_lowercase)
tweets['tweet_text'] = tweets['tweet_text'].apply(url_handling)
tweets['tweet_text'] = tweets['tweet_text'].apply(RT_handling)
tweets['tweet_text'] = tweets['tweet_text'].apply(dot_handling)
tweets['tweet_text'] = tweets['tweet_text'].apply(space_handling)
tweets['tweet_text'] = tweets['tweet_text'].apply(hashtag_handling)
tweets['tweet_text'] = tweets['tweet_text'].apply(smile)
tweets['tweet_text'] = tweets['tweet_text'].apply(laugh)
tweets['tweet_text'] = tweets['tweet_text'].apply(love)
tweets['tweet_text'] = tweets['tweet_text'].apply(wink)
tweets['tweet_text'] = tweets['tweet_text'].apply(sad)
tweets['tweet_text'] = tweets['tweet_text'].apply(cry)
#tweets['tweet_text'] = tweets['tweet_text'].apply(clap)
tweets['tweet_text'] = tweets['tweet_text'].apply(deEmojify)
```
